#  Netflix Data Analysis using Python

##  Project Overview

This project analyzes Netflix movies and TV shows using Python libraries such as Pandas, Matplotlib, and Seaborn. The goal is to explore the dataset and discover insights related to content types, genres, ratings, countries, release years, and Netflix growth trends.

##  Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

pandas → data handling

numpy → numerical operations

matplotlib → plotting graphs

seaborn → attractive visualizations

##  Loading Dataset

```python
df = pd.read_csv(
    '/kaggle/input/datasets/shivamb/netflix-shows/netflix_titles.csv'
)
```

Loads the Netflix dataset into a pandas DataFrame.

##  Checking Missing Values

```python
df.isnull().sum()
```

Checks how many missing values exist in each column.

##  Data Cleaning

```python
df.drop_duplicates(inplace=True)
```

Removes duplicate rows

```python
df.fillna('Unknown', inplace=True)
```

Fills missing values with "Unknown"

##  Movies vs TV Shows

```python
df['type'].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")

plt.show()
```

Shows the number of:

-Movies

-TV Shows

available on Netflix.

##  Top Countries Producing Netflix Content

```python
top_countries = df['country'].value_counts().head(10)

top_countries
```

```python
plt.figure(figsize=(10,5))

top_countries.plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")

plt.xlabel("Country")

plt.ylabel("Count")

plt.show()
```

Displays countries producing the most Netflix content.

##  Most Popular Genres

```python
top_genres = df['listed_in'].value_counts().head(10)

top_genres
```

```python
plt.figure(figsize=(12,5))

top_genres.plot(kind='bar')

plt.title("Most Popular Genres on Netflix")

plt.xlabel("Genre")

plt.ylabel("Count")

plt.show()
```

Analyzes most common genres such as:

-Drama

-Comedy

-Action

##  Ratings Analysis

```python
df['rating'].value_counts()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)

plt.title("Netflix Content Ratings")

plt.show()
```

Shows most common Netflix ratings:

TV-MA

PG-13

TV-14 etc.

##  Content Growth Over Years

```python
content_per_year = df['release_year'].value_counts().sort_index()

content_per_year
```

```python
plt.figure(figsize=(14,6))

content_per_year.plot()

plt.title("Netflix Content Growth Over Years")

plt.xlabel("Year")

plt.ylabel("Number of Titles")

plt.show()
```

Shows how Netflix content increased over time.

##  Movie Duration Analysis

```python
movies = df[df['type'] == 'Movie']

movies['duration'].head()
```

Analyzes movie durations available on Netflix.

##  Top Directors on Netflix

```python
top_directors = df['director'].value_counts().head(10)

top_directors
```

```python
plt.figure(figsize=(10,5))

top_directors.plot(kind='bar')

plt.title("Top Directors on Netflix")

plt.xlabel("Director")

plt.ylabel("Number of Titles")

plt.show()
```

Shows directors with the highest number of titles.

##  Most Frequent Actors

```python
top_cast = df['cast'].value_counts().head(10)

top_cast
```

Displays actors appearing most frequently.

##  Netflix Content Added Per Year

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,5))

df['year_added'].value_counts().sort_index().plot()

plt.title("Content Added to Netflix Per Year")

plt.xlabel("Year")

plt.ylabel("Count")

plt.show()
```

Shows how rapidly Netflix expanded its library.

##  Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```

Shows correlation between numerical features.

##  Key Insights

- Movies are more common than TV Shows on Netflix
- The United States produces the highest amount of Netflix content
- Drama is one of the most popular genres
- Netflix content increased rapidly after 2015
- TV-MA is the most common content rating
